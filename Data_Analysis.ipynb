```python
import pandas as pd
import numpy as np
from pandas import DataFrame,Series
```

```python
df_left_hr=DataFrame({'key1':['SF','SF','LA','LA'],
                      'key_2': [10,20,30,40],
                      'data_set': np.arange(4)})
```

```python
df1=DataFrame(np.arange(3),columns=['Mum'])
df2=DataFrame(np.arange(5),columns=['Del'])
df1.join(df2)
```

```python
df_left_hr
```

```python
#
df_right_hr=DataFrame(np.arange(8).reshape(4,2),index=[['LA','LA','SF',"SF"],[10,20,40,30]],
                                                        columns=['col1','col2'])
```

```python
df_right_hr
```

```python
#Index heirarchy
pd.merge(df_left_hr,df_right_hr,left_on=['key1','key_2'],right_index=True)
```

```python
#Concatenate in Numpy
arr1=np.arange(9).reshape(3,3)
np.concatenate([arr1,arr1],axis=1)
```

```python
#Concatenate in Pandas
ser1=Series([1,2,3], index=['T','U','V'])
ser2=Series([3,4],index=['X','Y'])
pd.concat([ser1,ser2],axis=1)
```

```python
pd.concat([ser1,ser2],keys=['cat1','cat2'])
```

```python
dframe1=DataFrame(np.random.randn(4,3),columns=['X','Y','Z'])
dframe2=DataFrame(np.random.randn(3,3),columns=['Y','Q','X'])
pd.concat([dframe1,dframe2])
```
